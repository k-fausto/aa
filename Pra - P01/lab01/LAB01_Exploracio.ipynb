<h1>Problema 1: BFS (2,5p)</h1>
<p>
    Implementeu l'algorisme <b>BFS</b>. Donat un graf de la llibreria networX, la funció <b>bfs()</b> ha de retornar les caselles accessibles des del paràmetre <b>origin</b>, incloent-hi aquesta, en l'ordre d'exploració. És a dir, la llista retornada sempre inclourà, com a mínim, el node orígen.<br>
    Recordeu que és necessari <b>no canviar l'estructura de la funció</b>. No canvieu ni el nom de la funció, ni el nom dels paràmetres d'entrada ni els de sortida. <br>
</p>    
    
</div>

````python
def bfs(G, origin):
    """
    Params
    ======    
    :G: Graf. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    
    Returns
    =======
    :visited: El conjunt de nodes visitats durant l'exploració BFS en l'ordre en que han estat explorats
    """
    visited = []

    # Comenzamos en el nodo origen y lo añadimos a la cola.
    queue = [origin]

    # Mientras haya elementos en la cola, seguiremos iterando.
    while queue:

        # Tomamos el primer elemento de la cola.
        current = queue.pop(0)

        # Si no lo hemos visitado, lo añadimos a la lista de visitados.
        # Luego, añadimos todos sus vecinos a la cola y continuamos iterando.
        if current not in visited:

            visited.append(current)

            for nei in G.neighbors(current):
                queue.append(nei)
    
    return visited

# Como comentario a pie de página, el código que hay en la presentación A002 del DFS añade los vecinos aunque el nodo ya haya sido visitado.
# ¿No es eso redundante? Si ya lo hemos visitado, ya añadimos sus vecinos. Estaríamos añadiendo gr(n) comprobaciones innecesarias.
````

<h1>Problema 2: Holes (2,5p)</h1>
<p>
    Implementeu una funció anomenada <b>holes()</b> que, donat un graf, un node orígen, un node destí, i un diccionari de nodes amb penalitzacions, trobi el camí més curt entre orígen i destí.<br>
    Si el camí passa per un dels nodes del diccionari, el cost de visitar el node serà la penalització en comptes de ser cost 1.<br>
    Heu de basar la vosta solució en l'algorisme <b>Dijkstra</b>.
</p>    
    
</div>

````python
# Implementación de Dijkstra.
import heapq as hq
import numpy as np

def dijkstra(G, origin):
    
    cost_to = {} # Siempre seguiré la estructura: cost_to[A] = (B, cost), con B siendo el previo y cost el costo. 
    #Inicializamos la lista de distancias a infinitas todas con el previo siendo nulo.
    for node in G.nodes():
        cost_to[node] = (None, np.inf)

    # Para el origen, definimos distancia cero.
    cost_to[origin] = (None, 0)

    # Crearemos el heap con el que procesaremos los nodos, poniendo en primer lugar el origen con su costo.
    heap = []
    hq.heapify(heap) # Siempre seguiré la estructura para los elementos del heap: (Coste al siguiente nodo, nodo actual).
    hq.heappush(heap, (0, origin))

    while heap:
        # Desempaquetamos el coste hasta el nodo actual, y el nodo actual.
        cost_of_path, current_node = hq.heappop(heap)

        # Si lo hemos visitado, ya tendremos su distancia mínima y continuamos. En caso contrario, lo añadimos a visitados.
        if current_node not in visited:

            visited.add(current_node)

            # Exploramos cada vecino del nodo actual. Si lo hemos visitado, ya tiene coste mínimo. 
            # En caso contrario, comprobamos si tenemos un nuevo mínimo
            for nei in G.neighbors(current_node):
                if nei not in visited:
                    weight = 1 # Peso arbitrario entre el actual y su vecino.
                    alternative_path = cost_of_path + weight 

                    # Si hemos encontrado un nuevo camino más corto, actualizamos cost_to.
                    if alternative_path < cost_to[nei][1]:
                        cost_to[nei] = (current_node, alternative_path)
                        hq.heappush(heap, (alternative_path, nei))

    return cost_to #Retorno el diccionario con el coste mínimo desde origen a todos los otros nodos del grafo.
````

````python
import heapq as hq
import numpy as np

def holes(G, origin, destination, holes_dct={}):
    """
    Params
    ======
    :G: Graf del qual en volem extreure el camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    :destination: Índex del node destí
    :holes_dct: Un diccionari del tipus {node: penalització}. Passar per cada node del diccionari té un cost diferent a 1.
    
    Returns
    =======
    :path: Una llista de nodes del camí més curt entre els nodes 'origin' i 'destination' (ambdós inclosos).
    :cost: Un enter amb el cost de recórrer el camí, incloent-hi les penalitzacions.
    """
    
    path = []
    cost = 0
    cost_to = {}
    visited = set()

    # Comprobemos que sea posible hacer un camino desde origin hasta destination.
    if destination not in bfs(G, origin):
        return [], -1

    # Creamos el diccionario con distancias mínimas a todos los nodos.
    for node in G.nodes():
        cost_to[node] = (None, np.inf) # Seguiré la estructura de cost_to[current] = (previous, cost)
        
    cost_to[origin] = (None, 0) #Puesto que comenzamos en el nodo de origen, no tiene previo y su coste es cero.

    # Creemos ahora el heap.
    heap = []
    hq.heapify(heap)
    hq.heappush(heap, (0, origin)) # Para que el heap pueda ordenar en base al coste, seguiré la estructura (coste, current).

    # Y utilicemos el heap para explorar todo el grafo con costes mínimos.
    while heap:

        current_cost, current_node = hq.heappop(heap) # Obtengo el siguiente camino de coste menor.

        # Si no hemos visitado el nodo, lo añadimos a visitados y vemos sus vecinos.
        if current_node not in visited:
            visited.add(current_node)

            # Para cada vecino vamos a calcular un coste alternativo desde el nodo actual.
            for nei in G.neighbors(current_node):

                # Recordemos que si está en visited su coste ya es mínimo.
                if nei not in visited:
                    # Asumiremos peso 1 como dice le enunciado, excepto en los casos que sea un agujero.
                    weight = 1 if nei not in holes_dct else holes_dct[nei] # Si es un nodo con penalización, añadimos su penalización.
                    alternative_path = current_cost + weight

                    # Si el coste es menor, actualizamos y tiramos al heap.
                    if alternative_path < cost_to[nei][1]:
                        cost_to[nei] = (current_node, alternative_path)
                        hq.heappush(heap, (alternative_path, nei))

    # Al vaciar el heap, tendremos el diccionario cost_to con todos los costes a todos los nodos. Solo queda reconstruir el camino.

    path = []
    cost = cost_to[destination][1]
    # Aquí establezco un nodo destino. 
    # Como cost_to tiene siempre el coste al actual y el nodo previo, puedo reconstruir un camino óptimo hasta origen
    # pasando desde el destino hasta el origen en orden inverso.
    pointer_to_node = destination
    while pointer_to_node != origin:
        path.append(pointer_to_node)
        pointer_to_node = cost_to[pointer_to_node][0]

    # Invertimos para obtener el orden normal.
    path.reverse()

    return path, cost
````

````python
# Proveu el vostre algorisme!

maze = Maze(30, 10, seed=15)

# Obtenim el graph associat al laberint on cada node es una cel·la i tindrem una aresta si entre dues cel·les no existeix paret
G = maze.maze_graph()

A = 0
B = len(G.nodes)-1
color = (255, 0, 255)

h_dcts = [
    {}, # Sense penalització
    {8:20,72:2, 227:10},
    {8:15, 72:10, 227:9},
]
for holes_dct in h_dcts:
    maze.set_values(list(holes_dct.keys()), 1, color)
    path, cost = holes(G, A, B, holes_dct)

    maze.set_path(path)
    print(f"Cost total: {cost}")
    print('Longitud del camí:', len(path))
    maze.display_maze(plot_path=True, height=200)

    maze.set_values(list(holes_dct.keys()), 0, color)
````

<p>
    Implementeu una funció anomenada <b>checkpoint()</b> que, donat un graf, un node origen, un node destí i un node extra, trobi el camí més curt des del node origen fins el node destí passant pel node extra. Com en l'exercici anterior, tindrem un diccionari de punts on aplicarem una penalització.<br>

 << RESPOSTA AQUÍ >>
    
    Este algoritmo tiene usa holes una vez, de complejidad O(V+E*logV), y dfs, de complejidad O(V+E).
    La construcción de los caminos tiene, en el peor de los casos, O(V). Esto es mejor que aplicar dos veces holes.
    Por tanto checkpoint() tiene complejidad O(V+E*logV).
</div>

````python
'''
    Función auxiliar que funciona exactamente igual a holes, pero devuelve cost_to en vez de el coste y camino.
'''

def holes_aux(G, origin, holes_dct={}):
    """
    Params
    ======
    :G: Graf del qual en volem extreure el camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    :holes_dct: Un diccionari del tipus {node: penalització}. Passar per cada node del diccionari té un cost diferent a 1.
    
    Returns
    =======
    :cost_to: Lista de distancias desde el nodo origen hasta el resto.
    """

    cost_to = {}
    visited = set()
    
    for node in G.nodes():
        cost_to[node] = (None, np.inf) # Seguiré la estructura de cost_to[current] = (previous, cost)
        
    cost_to[origin] = (None, 0) #Puesto que comenzamos en el nodo de origen, no tiene previo y su coste es cero.

    # Creemos ahora el heap.
    heap = []
    hq.heapify(heap)
    hq.heappush(heap, (0, origin)) # Para que el heap pueda ordenar en base al coste, seguiré la estructura (coste, current).

    # Y utilicemos el heap para explorar todo el grafo con costes mínimos.
    while heap:
        current_cost, current_node = hq.heappop(heap) # Obtengo el siguiente camino de coste menor.

        if current_node in visited:
            continue

        visited.add(current_node)

        for nei in G.neighbors(current_node):

            if nei in visited:
                continue

            
            weight = 1 if nei not in holes_dct else holes_dct[nei] # Si es un nodo con penalización, añadimos su penalización.
            alternative_path = current_cost + weight

            if alternative_path < cost_to[nei][1]:
                cost_to[nei] = (current_node, alternative_path)
                hq.heappush(heap, (alternative_path, nei))

    # Al vaciar el heap, tendremos el diccionario cost_to con todos los costes a todos los nodos. Solo queda reconstruir el camino.

    return cost_to
````

````python
def checkpoint(G, origin, destination, extra, holes_dct={}):
    """
    Params
    ======
    :G: Graf del qual en volem extreure el camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    :destination: Índex del node destí
    :extra: Índex d'un node extra per on ha de passar el camí
    :holes_dct: Un diccionari del tipus {node: penalització}. Passar per cada node del diccionari té un cost diferent a 1.
    
    Returns
    =======
    :path: Una llista de nodes del camí més curt entre els nodes 'origin' i 'destination' que passa per 'extra'.
    :cost: Un enter amb el cost de recórrer el camí, incloent-hi les penalitzacions.
    """
    path = []
    cost = 0

    # Comprobemos que sea posible hacer un camino desde origin hasta destination.
    if destination not in bfs(G, origin):
        return [], -1
    
    # Obtenemos el diccionario con todos los costes desde extra. Este diccionario tiene a su vez la distancia mínima hasta origen
    # y la distancia mínima hasta destination, así como la información para construir el camino.
    # Esto es mejor que llamar dos holes.
    cost_to = holes_aux(G, extra, holes_dct)

    # Creemos el camino desde origen hasta extra.
    path_from_origin_to_extra = []
    cost_from_origin_to_extra = cost_to[origin][1]
    
    # Empezamos el origin, calculamos su previo, vemos el previo del previo...
    # Así hasta llegar a extra. Donde habremos creado el camino hasta extra en orden.
    pointer_to_node = origin
    while pointer_to_node != extra:
        path_from_origin_to_extra.append(pointer_to_node)
        pointer_to_node = cost_to[pointer_to_node][0]


    # camino desde extra hasta destination.
    path_from_extra_to_destination = []
    cost_from_extra_to_destination = cost_to[destination][1]
    pointer_to_node = destination
    while pointer_to_node != extra:
        pointer_to_node = cost_to[pointer_to_node][0]
        path_from_extra_to_destination.append(pointer_to_node)
    
    #La diferencia es que en este caso, los elementos se añaden en orden inverso y por tanto tenemos que arreglarlo.
    path_from_extra_to_destination.reverse()
    
    path = path_from_origin_to_extra + path_from_extra_to_destination # Sumamos los dos caminos.
    cost = cost_from_origin_to_extra + cost_from_extra_to_destination # Sumamos los costos.
    
    #Corregimos el hecho de que nos falta añadir el último.
    path.append(destination)

    return path, cost
````

````python
# Proveu el vostre algorisme!

# Creem un laberint i n'extraiem el graph
maze = Maze(6,6, seed=17)
G = maze.maze_graph()

# Definim l'inici, el fi, un node extra i la llista amb penalització
A = 30
B = len(G.nodes)-1
E = 8

h_dcts = [
    {}, # Sense penalització
    {13:3, 28:5},
    {13:10, 28:20},
]
for holes_dct in h_dcts:
    # La funció set_value ens permet 'activar' un node per a que surti dibuixat al laberint.
    maze.set_value(E, 1)                         # Checkpoint
    maze.set_values(list(holes_dct.keys()), 1, (255,0,255))  # Penalitzacions

    path, cost = checkpoint(G, A, B, E, holes_dct)
    print(f'El cost del camí més curt que passa per E és: {cost}')
    print('La distàcia del camí més curt que passa per E és:', len(path))
    print(f'El camí és: {path}')
    maze.set_path(path)
    maze.display_maze(plot_path=True, height=200)
    
    maze.set_value(E, 0)                         # Checkpoint
    maze.set_values(list(holes_dct.keys()), 0, (255,0,255))  # Penalitzacions
````

 Esto es porque calcularemos n! veces path_calculator, que tiene complejidad O(V * n).
 Y haremos n veces holes, que tiene complejidad O(V + E * log(v))
 Sacando factor común n, sale  la complejidad descrita anteriormente.
    
</div>

````python
# Esta función, dada un diccionario con los costos a todos los nodos, un nodo origen y un nodo destino, genere
# el camino de menor coste hasta el nodo anterior a destination y su costo.

def path_constructor(cost_to, origin, destination):
    path = []
    cost = cost_to[destination][1]
    pointer_node = destination
    while pointer_node != origin:
        path.append(pointer_node)
        pointer_node = cost_to[pointer_node][0]
        
    return path, cost

# Creemos una función auxiliar para producir un camino y costo dados los diccionarios de distancia, una permutación, un nodo origen
# y un nodo de destino.

def path_calculator(list_of_costs, permutation, origin, destination, holes_dct):
  
    # Asigno un nodo puntero. Este será originalmente origin.
    pointer_node = origin
    path = []
    cost = 0

    for elem in permutation:
        # Tomamos el primer elemento de la permutación. Calcularemos el mini camino desde el nodo puntero hasta el primer elemento de la permutación.
        # Calcularemos el camino y distancia, y añadiremos el coste. Así, iremos construyendo de nodo a nodo el camino.
        index, node = elem
        cost += list_of_costs[index][node][1]
        
        sub_path, subcost = path_constructor(list_of_costs[index], node, pointer_node)
        path += sub_path
        cost += subcost
        pointer_node = node # Movemos el puntero al siguiente nodo.

    # Corrección.
    # Destination debe ser el último y por tanto lo añadimos de esta manera.
    # Además, path_constructor no añade el último nodo, así que debe añadirse manualmente destination.
    sub_path, subcost = path_constructor(holes_aux(G, destination, holes_dct), destination, pointer_node)
    path += sub_path
    cost += subcost - 1
    cost += 1 if destination not in holes_dct else holes_dct[destination]
    path.append(destination)

    return path, cost

    
````

````python
import itertools as it
import numpy as np

def checkpoints_list(G, origin, destination, extras, holes_dct={}):
    """
    Params
    ======
    :G: Graf del qual en volem extreure el camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    :destination: Índex del node destí
    :extras: Llista d'índexs de nodes per on ha de passar el camí.
    :holes_dct: Un diccionari del tipus {node: penalització}. Passar per cada node del diccionari té un cost diferent a 1.
    
    Returns
    =======
    :path: Una llista de nodes del camí més curt entre els nodes 'origin' i 'destination' que passa per tots els nodes 'extras'.
    :cost: Un enter amb el cost de recórrer el camí, incloent-hi les penalitzacions.
    """

    path = []
    cost = 0


    reachable_nodes = bfs(G, origin)      

    # Creo una copia de extras, porque prefiero no editar la lista original.
    i = 0
    required_nodes = [] # Seguiré la estructura (índice, nodo).
    for extra in extras:
        required_nodes.append((i, extra))
        i+=1

    # Comprobamos que exista un camino.
    for required in required_nodes:
        if required[1] not in reachable_nodes:
            return [], -1
        
    # Creo la lista de costes. Para cada extra, tenemos su diccionario de costes asociado.
    list_of_costs = []
    for required in required_nodes:
        list_of_costs.append(holes_aux(G, required[1], holes_dct))

    #Vamos a ver cada permutación, calcular el coste total de sus caminos, y quedarnos con la mejor.
    min_cost = (np.inf, None) # Seguiré la estructura, (costo, permutación), de forma que pueda recuperar la permutación mínima.
    
    for permutation in it.permutations(required_nodes): # Estas permutaciones son de la forma [(índice, nodo_extra), ...]

        permutation_cost = 0
        # Usamos la función auxiliar para calcular el camino y coste. Si es mejor, lo haremos el mínimo. En caso contrario, continuamos iterando.
        permutation_path, permutation_cost = path_calculator(list_of_costs, permutation, origin, destination, holes_dct)

        if permutation_cost < min_cost[0]:
            min_cost = (permutation_cost, permutation_path)

    cost = min_cost[0]
    path = min_cost[1]

    return path, cost
````
