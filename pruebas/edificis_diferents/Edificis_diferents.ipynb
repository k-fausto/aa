<h1>Modificació: Edificis diferents</h1>

Donat el problema original es modifica per a tenir un conjunt d'edificis, tots ells de diferent alçada. Si la graella és $N\times N$ tenim edificis d'alçades entre $1$ i $N^2$. <b>En tota la graella no es pot repetir cap alçada.</b> Al igual que al problema original, els números a l'exterior indiquen quants edificis són visibles des d'aquell costat, considerant que un edifici més alt amaga els situats al darrere.


<b>Restriccions:</b> 

<ul>
    <li> Cada alçada entre $1$ i $N^2$ s'utilitza exactament una vegada en tota la graella.
    </li>
    <li>La regla de visibilitat és la mateixa que a la versió original.</li>
</ul>
</p>
    <p>
    L’objectiu és trobar una configuració d’edificis (assignació d’alçades entre $1$ i $N^2$ per les $N^2$ caselles que satisfaci les condicions de visibilitat i la no repetició d’alçades. 
    </p>
    
</div>

La funció **format_sky** us ajudarà a visualitzar el tauler del problema.

Veieu un exemple de com utilitzar aquesta funció:


````python
import numpy as np
from IPython.display import clear_output
from utils_edificis_diferents import format_sky
import time

# Amb aquesta funció es mostren totes les caselles del tauler
# Les caselles que contenen un zero es mostren buides

# Creem unes restriccions i un tauler 
top, bottom, left, right = [0,0,0], [3,0,0], [0,0,0], [0,0,1]
grid = np.zeros((len(left), len(top)), dtype='int')

print('Tauler:')
print(format_sky(grid, top, bottom, left, right))
````

La funció **delayed_print** us permet bloquejar temporalment un 'print' per si voleu veure les accions que fa a cada pas el vostre algorisme.<br>
No heu d'utilitzar-la a l'entrega ja que el programa tardarà molt més del necessari, utilitzeu-la només mentre programeu la solució.

````python
def delayed_print(grid, top, bottom, left, right, sleep_time=0.02):
    """
    Funció auxiliar que bloqueja temporalment la funció 'print' per veure l'evolució de l'algorisme.
    """
    print(format_sky(grid, top, bottom, left, right))
    clear_output(wait=True)
    time.sleep(sleep_time)
````

````python
###
# LES VOSTRES FUNCIONS EXTRES AQUÍ
###

def satisfies(grid, x, y, num, top, bottom, left, right, visited):
    """
    Donat un tauler, un punt amb coordenades (x,y) i un nombre 'num', comprova si és possible assignar-lo a la posició (x,y) 
    donades les restriccions del problema.
    
    Params
    ------
    :grid: Una matriu de numpy
    :x,y: Un punt de la matriu on volem posar el nombre 'num'
    :num: Un nombre enter
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    
    Returns
    -------
    :b: Un boleà True/False depenent de si el nombre 'num' pot ser col·locat a la casella (x,y) o no.
    """
    
    # EL TEU CODI AQUÍ
    n = len(grid)

    # Comprobamos que ningún elemento en la fila o columna tenga el mismo valor.

    # Números repetidos.
    if num in visited:
        return False

    # Si veo k elementos, debo ver k+3, y solo quedan 2 casillas disponibles, entonces es seguro que no podré cumplir la condición.
    # Entonces, comprobaré los visibles hasta ahora, la cantidad de casillas restantes, y comprobaré mi condición.

    # Para top: contamos la cantidad de edificios visibles. 
    # Si las casillas disponibles son menos que los que nos faltan por ver, retornamos false, pues no podremos cumplir el requisito.
    
    # Vemos qué sería visible si llegamos a poner un edificio de altura num en el lugar.
    grid[x, y] = num

    # Comprobación para top.
    count = 0
    current_max = 0
    remaining_empty = n - np.count_nonzero(grid[:, y])

    for i in range(n):
        if current_max < grid[i, y]:
            count += 1
            current_max = grid[i, y] 

    # Contaremos los no-ceros y restaremos a n la cantidad de no-ceros.

    # Si tenemos un número fuera de la casilla, tenemos que ver que, queden más o igual cantidad de
    # de casillas disponibles que la diferencia entre las que necesitamos ver y las que vemos.
    # Si no es el caso, retornamos false. La lógica es idéntica con el resto de casos.
    if top[y] != 0 and remaining_empty < max(0, top[y] - count):
        grid[x, y] = 0
        return False 

    # Comprobación para bottom.
    count = 0
    current_max = 0

    for i in range(n-1, -1, -1):
        if current_max < grid[i, y]:
            count += 1
            current_max = grid[i, y] 

    if bottom[y] != 0 and remaining_empty < max(0, bottom[y] - count):
        grid[x, y] = 0
        return False 

    # Comprobación para left.
    count = 0
    current_max = 0
    remaining_empty = n - np.count_nonzero(grid[x, :])

    for j in range(n):
        if current_max < grid[x, j]:
            count += 1
            current_max = grid[x, j] 

    if left[x] != 0 and remaining_empty < max(0, left[x] - count):
        grid[x, y] = 0
        return False 

    # Comprobación para right.
    count = 0
    current_max = 0
    for j in range(n-1, -1, -1):
        if current_max < grid[x, j]:
            count += 1
            current_max = grid[x, j] 
    if right[x] != 0 and remaining_empty < max(0, right[x] - count):
        grid[x, y] = 0
        return False 

    grid[x, y] = 0
    return True
    
def check_cols_and_rows(grid, top, bottom, left, right, n): # Esta función auxiliar comprueba que las condiciones del problema se cumplan. (vistas)
    # Retorna True si es valida, false en caso que no.
    # Comprobamos que cada visión sea correcta, para ver si la solución es correcta.

    for i in range(n):

        # Top.
        count = 0
        current_max = 0
        for j in range(n):
            if current_max < grid[j, i]:
                count += 1
                current_max = grid[j, i] 

        if top[i] != 0 and count != top[i]:
            return False

        # Bottom.
        count = 0
        current_max = 0
        for j in range(n):
            if current_max < grid[n-1-j, i]:
                count += 1
                current_max = grid[n-1-j, i] 

        if bottom[i] != 0 and count != bottom[i]:
            return False
        
        # Left.
        count = 0
        current_max = 0
        for j in range(n):
            if current_max < grid[i, j]:
                count += 1
                current_max = grid[i, j] 
        
        if left[i] != 0 and count != left[i]:
            return False

        # Right.
        count = 0
        current_max = 0
        for j in range(n):
            if current_max < grid[i, n-1-j]:
                count += 1
                current_max = grid[i, n-1-j] 

        if right[i] != 0 and count != right[i]:
            return False

    #Si pasa todas las comprobaciones, devolvemos true.
    return True

def skyscrapper_backtracking_r(grid, top, bottom, left, right, visited):  
    """
    Funció que implementa l'algorisme de backtracking
    
    Params
    ------
    :grid: Una matriu de tipus numpy
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    
    Returns
    -------
    :grid: La matriu plena amb la solució correcta o bé False en cas que no tingui solució.    
    """
    
    # NO esborreu aquestes dues línies, serveien per saber quants cops es crida aquesta funció
    # ----------------------------------------------------------------------------------------
    global num_recursive_calls 
    num_recursive_calls += 1
    # ----------------------------------------------------------------------------------------
    
    # EL TEU CODI AQUÍ
    
    n = len(grid)

    # Comprobamos que las filas y columnas estén bien. 
    if np.count_nonzero(grid) == n*n:
        # Si cumple las condiciones del problema y tiene n*n elementos, la solución es correcta.
        if check_cols_and_rows(grid, top, bottom, left, right, n):
            return grid.tolist()
        # Si la comprobación falla, retornamos false.
        return False

    for j in range(n):
        for i in range(n):

            # Si ya hemos fijado el valor, saltamos la columna.
            if grid[i, j] == 0:

            # Para cada casilla vacía, veremos cada valor posible.
                for k in range(1, n*n+1):

                    if satisfies(grid, i, j, k, top, bottom, left, right, visited): # Si es un movimiento válido:
                        # Aplicamos el movimiento, hacemos backtracking, y si la solución es válida, retornamos el grid.

                        grid[i, j] = k
                        visited.add(k)
                        ok = skyscrapper_backtracking_r(grid, top, bottom, left, right, visited)
                        if ok is not False:
                            return ok
                        
                        # Si la solución no era válida, deshacemos el movimiento.
                        grid[i, j] = 0
                        visited.remove(k)

                # Si ningún valor es válido para esta casilla, retornamos False, pues no es una recursión con solución.
                return False

    return False

def skyscrapper_backtracking(grid, top, bottom, left, right):
    visited = set()
    return skyscrapper_backtracking_r(grid, top, bottom, left, right, visited)
    
        

#
# NO MODIFIQUEU AQUESTA FUNCIÓ, PODEU MODIFICAR LES ALTRES I/O AFEGIR MÉS
#
from utils_edificis_diferents import gradable

@gradable   
def skyscrapper(top, bottom, left, right):
    """
    Funció principal del problema. Rep quatre llistes corresponents als nombres que hi ha fora del tauler.
    
    Params
    ------
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    """
    
    # Inicialitzem una matriu de zeros
    grid = np.zeros((len(left), len(top)), dtype='int')
    
    # Cridem a la funció que soluciona el problema mitjançant backtracking
    sol = skyscrapper_backtracking(grid, top, bottom, left, right)   
    
    return sol
````
