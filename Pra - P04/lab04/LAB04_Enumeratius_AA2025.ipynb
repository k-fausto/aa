La funció **format_sky** us ajudarà a visualitzar el tauler del problema.

````python
import numpy as np
from IPython.display import clear_output
import time

def format_sky(board, top, bottom, left, right):
    """
    Funció auxiliar per mostrar la matriu del Problema 1 en forma de graella.
    """
    _str = ""
    
    # Files
    for v in top:
        _str+="   "+str(v)
    _str = "  " + _str + "\n"

    for idx, i in enumerate(board):
        _str+= "   +"+ ("-"*((board.shape[1]*4)-1)) +"+\n"
        _str+= f" {left[idx]} | "
        
        # Columnes
        for j in i:   
            if j!=0:
                _str+= str(j)+" | "
            else:
                 _str+= "  | "
            
        _str+= f"{right[idx]} \n"
    _str+= "   +"+ ("-"*((board.shape[1]*4)-1))+"+\n"
    
    _bot = ""
    for v in bottom:
        _bot+="   "+str(v)
    _str += "  " + _bot
    
    return _str.replace('0',' ')
````

Veieu un exemple de com utilitzar aquesta funció:


````python
# Amb aquesta funció es mostren totes les caselles del tauler
# Les caselles que contenen un zero es mostren buides

# Creem unes restriccions i un tauler 
top, bottom, left, right = [0,0,0], [3,0,0], [0,0,0], [0,0,1]
grid = np.zeros((len(left), len(top)), dtype='int')

print('Tauler:')
print(format_sky(grid, top, bottom, left, right))
````

La funció **delayed_print** us permet bloquejar temporalment un 'print' per si voleu veure les accions que fa a cada pas el vostre algorisme.<br>
No heu d'utilitzar-la a l'entrega ja que el programa tardarà molt més del necessari, utilitzeu-la només mentre programeu la solució.

````python
def delayed_print(grid, top, bottom, left, right, sleep_time=0.02):
    """
    Funció auxiliar que bloqueja temporalment la funció 'print' per veure l'evolució de l'algorisme.
    """
    print(format_sky(grid, top, bottom, left, right))
    clear_output(wait=True)
    time.sleep(sleep_time)
````

````python
###
# LES VOSTRES FUNCIONS EXTRES AQUÍ
###

def satisfies(grid, x, y, num, top, bottom, left, right):
    """
    Donat un tauler, un punt amb coordenades (x,y) i un nombre 'num', comprova si és possible assignar-lo a la posició (x,y) 
    donades les restriccions del problema.
    
    Params
    ------
    :grid: Una matriu de numpy
    :x,y: Un punt de la matriu on volem posar el nombre 'num'
    :num: Un nombre enter
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    
    Returns
    -------
    :b: Un boleà True/False depenent de si el nombre 'num' pot ser col·locat a la casella (x,y) o no.
    """
    
    # EL TEU CODI AQUÍ

    n = len(grid)

    # Comprobamos que ningún elemento en la fila o columna tenga el mismo valor.

    # Fila.
    for j in range(n):
        if grid[x, j] == num:
            return False
        
    # Columna.
    for i in range(n):
        if grid[i, y] == num:
            return False
        
    # Comprobemos ahora valores imposibles para cada top, bottom, left, right.
    # Por ejemplo, si queremos asignar un valor n, y el número fuera de la casilla fuese 1, 
    # la distancia de n al número fuera de la casilla no puede ser mayor que 1 (siendo este adyacente al número),
    # pues de lo contrario forzosamente veríamos dos elementos como mínimo.

    # Para top:
    #if num + top[y] >= n+2:
    #    if x < (num + top[y] - n):
    #        return False

    return True

def check_cols_and_rows(grid, top, bottom, left, right): # Esta función auxiliar comprueba que las condiciones del problema se cumplan. 
    # Retorna True si es valida, false en caso que no.

    n = len(grid)
    # Comprobamos que cada visión sea correcta, para ver si la solución es correcta.

    for i in range(n):


        # Left.
        count = 0
        current_max = 0
        for j in range(n):
            if current_max < grid[i, j]:
                count += 1
                current_max = grid[i, j] 
        
        if left[i] != 0 and count != left[i]:
            return False

        # Right.
        count = 0
        current_max = 0
        for j in range(n):
            if current_max < grid[i, n-1-j]:
                count += 1
                current_max = grid[i, n-1-j] 

        if right[i] != 0 and count != right[i]:
            return False

        # Top.
        count = 0
        current_max = 0
        for j in range(n):
            if current_max < grid[j, i]:
                count += 1
                current_max = grid[j, i] 

        if top[i] != 0 and count != top[i]:
            return False

        # Bottom.
        count = 0
        current_max = 0
        for j in range(n):
            if current_max < grid[n-1-j, i]:
                count += 1
                current_max = grid[n-1-j, i] 

        if bottom[i] != 0 and count != bottom[i]:
            return False


    #Si pasa todas las comprobaciones, devolvemos true.
    return True
    

def skyscrapper_backtracking(grid, top, bottom, left, right):  
    """
    Funció que implementa l'algorisme de backtracking
    
    Params
    ------
    :grid: Una matriu de tipus numpy
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    
    Returns
    -------
    :grid: La matriu plena amb la solució correcta o bé False en cas que no tingui solució.    
    """
    
    # NO esborreu aquestes dues línies, serveien per saber quants cops es crida aquesta funció
    # ----------------------------------------------------------------------------------------
    global num_recursive_calls 
    num_recursive_calls += 1
    # ----------------------------------------------------------------------------------------
    
    # EL TEU CODI AQUÍ

    n = len(grid)

    if np.count_nonzero(grid) == n*n:
        # Comprobamos que las filas y columnas estén bien.
        if check_cols_and_rows(grid, top, bottom, left, right):
            return grid
        # Si la comprobación falla, retornamos false.
        return False

    for i in range(n):
        for j in range(n):

            # Si ya hemos fijado el valor, saltamos la columna.
            if grid[i, j] == 0:
            # Para cada casilla vacía, veremos cada valor posible.
                for k in range(1, n+1):

                    if satisfies(grid, i, j, k, top, bottom, left, right): # Si es un movimiento válido:

                        # Aplicamos el movimiento, hacemos backtracking, y si la solución es válida, retornamos el grid.
                        # print(f"haré que en la casilla ({i}, {j}) esté el valor {k}")
                        # print(format_sky(grid, top, bottom, left, right))
                        grid[i, j] = k

                        ok = skyscrapper_backtracking(grid, top, bottom, left, right)

                        if ok is not False:
                            return ok
                        
                        # Si la solución no era válida, deshacemos el movimiento.
                        grid[i, j] = 0
                        # print("DESHAGO EL MOVIMIENTO Y EL TABLERO QUEDA ASÍ:")
                        # print(format_sky(grid, top, bottom, left, right))

                # Si ningún valor es válido para esta casilla, retornamos False.
                return False


    return False
        
    
def skyscrapper(top, bottom, left, right):
    """
    Funció principal del problema. Rep quatre llistes corresponents als nombres que hi ha fora del tauler.
    
    Params
    ------
    :top, bottom, left, right: Els nombres de fora del tauler de la part superior, inferior, esquerra i dreta, respectivament.
    """
    
    # Inicialitzem una matriu de zeros
    grid = np.zeros((len(left), len(top)), dtype='int')

    n = len(grid)

    # Trabajo preliminar:
    # Si un número fuera del tablero es 1, debemos poner el rascacielos más alto en frente de este número, de lo contrario vería más de un elemento.
    # Si un número fuera del tablero es n, con n la medida del tablero, la permutación queda unívocamente definida como 1, 2, ..., n.

    for i in range(n):
        # Manejo de top:
        if top[i] == n: 
            for j in range(n): grid[j, i] = j+1
        elif top[i] == 1: 
            grid[0, i] = n
            
        # Manejo de bottom:
        if bottom[i] == n: 
            for j in range(n): grid[j, i] = n-j
        elif bottom[i] == 1: 
            grid[n-1, i] = n

        # Manejo de left:
        if left[i] == n: 
            for j in range(n): grid[i, j] = j+1
        elif left[i] == 1: 
            grid[i, 0] = n

        # Manejo de top:
        if right[i] == n: 
            for j in range(n): grid[i, j] = n-j
        elif right[i] == 1: 
            grid[i, n-1] = n
    
    # Cridem a la funció que soluciona el problema mitjançant backtracking
    sol = skyscrapper_backtracking(grid, top, bottom, left, right)
    
    # Mostrem el resultat en el cas que trobem una solució o mostrem un error en cas contrari.
    if sol is not False:
        sol = sol.tolist()
        print(format_sky(np.array(sol), top, bottom, left, right))
    else:
        print('No solution found')
````

````python
num_recursive_calls = 0
t0 = time.perf_counter()
skyscrapper([0,0,0],
            [3,0,0],
            [0,0,0],
            [0,0,1])
t1 = time.perf_counter()
print('Crides recursives:', num_recursive_calls)
print(f'Temps: {t1-t0:.2f}s')
````
